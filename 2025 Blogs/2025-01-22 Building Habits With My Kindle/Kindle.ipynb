### Kindle.Devices.ReadingSession

```sql
%python
import pandas as pd
df_rs = pd.read_csv('/.../KindleDevicesReadingSession.csv')
```

### Take a look at example data

```sql

%python
from IPython.display import display, HTML
display(HTML(df_rs.head().to_html()))
```

```sql
%python
# Assuming your DataFrame is named 'df' and the column with milliseconds is named 'duration_ms'
df_rs['duration_minutes'] = df_rs['total_reading_millis'] / (1000 * 60)
df_rs['start_timestamp'] = pd.to_datetime(df_rs['start_timestamp'], errors='coerce')
df_rs = df_rs[df_rs['start_timestamp'].dt.year == 2025]
```

```sql
%python
# Define a dictionary where keys are the original column names and values are the new names
columns_to_keep = {
    'start_timestamp': 'start_timestamp',
    'end_timestamp': 'end_timestamp',
    'ASIN': 'ASIN',
    'duration_minutes': 'duration_minutes',
    'number_of_page_flips' : 'number_of_page_flips'
}

# Keep only the specified columns and rename them
df_rs = df_rs[list(columns_to_keep.keys())].rename(columns=columns_to_keep)

# Display the result
display(HTML(df_rs.head().to_html()))
```

```sql
%python
df_titled = pd.merge(df_rs, df_br, on='ASIN', how='left')

# Sort the DataFrame by 'Product Name' in ascending order
df_titled = df_titled.sort_values('start_timestamp', ascending=True)
```

```sql
%python
# Display the result
display(HTML(df_titled.head().to_html()))
```

### Pages Turned Per Day


```sql
%python
import matplotlib.pyplot as plt

# Convert start_timestamp to datetime and extract the date
df_titled['date'] = pd.to_datetime(df_titled['start_timestamp']).dt.date
```

```sql
%python
# Group by date and sum the number of page flips
daily_page_flips = df_titled.groupby('date')['number_of_page_flips'].sum().reset_index()

# Sort by date
daily_page_flips = daily_page_flips.sort_values('date')

# Create the line chart
plt.figure(figsize=(12,6))
plt.plot(daily_page_flips['date'], daily_page_flips['number_of_page_flips'])
plt.title('Number of Page Flips per Day')
plt.xlabel('Date')
plt.ylabel('Number of Page Flips')
plt.xticks(rotation=45)
plt.tight_layout()
```

### Non Reading Days


```sql
%python
# Create a continuous date range
date_range = pd.date_range(start=daily_page_flips['date'].min(), end=daily_page_flips['date'].max())

# Reindex the DataFrame to include all dates, filling missing values with 0
daily_page_flips = daily_page_flips.set_index('date').reindex(date_range, fill_value=0).reset_index()
daily_page_flips = daily_page_flips.rename(columns={'index': 'date'})

# Calculate the average number of page flips (excluding days with no reading)
average_page_flips = daily_page_flips[daily_page_flips['number_of_page_flips'] > 0]['number_of_page_flips'].mean()

# Create the line chart
plt.figure(figsize=(12,6))

# Plot all days with a blue line
plt.plot(daily_page_flips['date'], daily_page_flips['number_of_page_flips'], color='blue', label='Reading activity')

# Plot red circles for days without reading activity
no_reading_days = daily_page_flips[daily_page_flips['number_of_page_flips'] == 0]
plt.scatter(no_reading_days['date'], no_reading_days['number_of_page_flips'], color='red', marker='o', s=30, label='No reading')

# Add average page flips line
plt.axhline(y=average_page_flips, color='green', linestyle='--', label=f'Average ({average_page_flips:.2f} flips)')

plt.title('Number of Page Flips per Day')
plt.xlabel('Date')
plt.ylabel('Number of Page Flips')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
```

### Time reading each day


```sql
%python
# Group by date and sum the duration_minutes
daily_reading_duration = df_titled.groupby('date')['duration_minutes'].sum().reset_index()

# Create a continuous date range
date_range = pd.date_range(start=daily_reading_duration['date'].min(), end=daily_reading_duration['date'].max())

# Reindex the DataFrame to include all dates, filling missing values with 0
daily_reading_duration = daily_reading_duration.set_index('date').reindex(date_range, fill_value=0).reset_index()
daily_reading_duration = daily_reading_duration.rename(columns={'index': 'date'})

# Calculate the average reading duration (excluding days with no reading)
average_duration = daily_reading_duration[daily_reading_duration['duration_minutes'] > 0]['duration_minutes'].mean()

# Create the line chart
plt.figure(figsize=(12,6))

# Plot all days with a blue line
plt.plot(daily_reading_duration['date'], daily_reading_duration['duration_minutes'], color='blue', label='Reading duration')

# Plot red circles for days without reading activity
no_reading_days = daily_reading_duration[daily_reading_duration['duration_minutes'] == 0]
plt.scatter(no_reading_days['date'], no_reading_days['duration_minutes'], color='red', marker='o', s=30, label='No reading')

# Add average duration line
plt.axhline(y=average_duration, color='green', linestyle='--', label=f'Average ({average_duration:.2f} min)')

plt.title('Total Reading Duration per Day')
plt.xlabel('Date')
plt.ylabel('Reading Duration (minutes)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
```
